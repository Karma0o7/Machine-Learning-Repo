<h2>Relationships between items</h2>
<h3>Corelations Graph is used to visualize relationship between features</h3>
The scientific term characterizing the "relationship" between features is correlation.
So here 1 means total correlation and as expected the diagonal is all ones because a feature is totally correlated with its self. Also, the lower the number, the less correlated are the features.

Here we need to consider the feature-to-feature correlations and the outcome-to-feature correlations.

Higher correlations Between features: higher correlations mean we can drop one of them. 

However, high correlation between a feature and the outcome means that the feature is important and holds a lot of information.

```python
import pandas as pd
import seaborn as sns
from pylab import rcParams
import matplotlib.pyplot as plt

def plot_correlation(data):
    '''
    plot correlation's matrix to explore dependency between features 
    '''
    # set the size of the plot
    rcParams['figure.figsize'] = 20,20
    fig = plt.figure()

    # plot the correlation's matrix
    sns.heatmap(data.corr(), annot=True, fmt=".2f")

    # show the plot
    plt.show()
    fig.savefig('correlation.png')

# load the data
path="https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-ML0101EN-SkillsNetwork/labs/Module%203/data/teleCust1000t.csv"
df = pd.read_csv(path)

# Call the plot_correlation function
plot_correlation(df)
```
